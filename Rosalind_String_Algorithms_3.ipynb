```python
#ROSALIND STRING ALGORITHMS 3

#Complementing a Strand of DNA

#Instructions:
#In DNA strings, symbols 'A' and 'T' are complements of each other, as are 'C' and 'G'.

#The reverse complement of a DNA string s is the string sc formed by reversing the symbols of s,
#then taking the complement of each symbol (e.g., the reverse complement of "GTCA" is "TGAC").

#Given: A DNA string s of length at most 1000 bp.

#Return: The reverse complement sc of s.
```

```python
dna="AAAACCCGGT"

complement_list = [] # Initialize an empty list to store all
#complementary nucleotides
for nucleo in dna: # Inicialmente estaba iterando dentro del for
#pero esto es una mejor solución
#mis líneas originales eran
#nucleo="" *Este pasa a definirse en la iteración
#new="" *Este era innecesario si de agregaba directamente la nueva
#letra a la lista en vez de guardarla en una nueva dirección
#complement=[] *Se define por fuera para evitar huardados por
#separado aka. verticales
  if nucleo=="A":
    complement_list.append("T")
  elif nucleo=="T":
    complement_list.append("A")
  elif nucleo=="G":
    complement_list.append("C")
  elif nucleo=="C":
    complement_list.append("G")

complement = "".join(complement_list) # Para que queden ena línea
#hacer la unión fuera del for
reverse=complement[::-1] #reverso complemento por slicing (voltea
#la data)
print (reverse)
```

```python
#El código en limpio

dna=open('/content/drive/MyDrive/Rosalind/rosalind_revc.txt').read().strip()

#Initialize an empty list to store all complementary nucleotides
#complement_list=[] *It's defined outside to avoid storing them separately
#This prevents creating separate vertical outputs and saves data together

complement_list = []

#Define the list outside the loop to store the full sequence
#Keeping this outside prevents resetting the list on each iteration
#This cycle compares every position in the original string with individual nucleotides
#Inside every if it defines a line complement_list.append(n)
#This command adds the complementary nucleotide of the original string position
#to the new list.

for nucleo in dna:
  if nucleo=="A":
    complement_list.append("T")
  elif nucleo=="T":
    complement_list.append("A")
  elif nucleo=="G":
    complement_list.append("C")
  elif nucleo=="C":
    complement_list.append("G")

# "".join(list) puts together every list item separated by "" (nothing in this scenario)
#into a single string

complement = "".join(complement_list)

#Slicing reads complement from end to start, inverting the sequence direction
#and assigning this new reading to the variable reverse

reverse=complement[::-1]
print (reverse)
```

```python
#A more informatic solution
#If you undersand the mechanics behind the complex code, you'll be better
#at finding simple solutions

dna = open('/content/drive/MyDrive/Rosalind/rosalind_revc.txt').read().strip()

# maketrans creates a direct sustitution map
tabla_sustitucion = str.maketrans("ATCG", "TAGC")

# translate applies the map and [::-1] inverts
reverse = dna.translate(tabla_sustitucion)[::-1]

print(reverse)
```
