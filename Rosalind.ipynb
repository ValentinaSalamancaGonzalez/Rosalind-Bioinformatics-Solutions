````python
def solve_gc_content(filename):
    with open(filename, 'r') as file:
        lines = file.read().splitlines()

    seq_dict = {}
    current_id = ""

    # 1. Parse FASTA file
    for line in lines:
        if line.startswith(">"):
            current_id = line[1:]
            seq_dict[current_id] = ""
        else:
            seq_dict[current_id] += line

    max_id = ""
    max_gc = -1.0

    # 2. Calculate GC content
    for seq_id, seq in seq_dict.items():
        if not seq:
            continue
        # Formula: (Count of G and C / Total length) * 100
        gc_count = seq.count('G') + seq.count('C')
        gc_percent = (gc_count / len(seq)) * 100

        # 3. Find the highest
        if gc_percent > max_gc:
            max_gc = gc_percent
            max_id = seq_id

    print(max_id)
    print(f"{max_gc:.6f}")

# Call the function with your input file
solve_gc_content('/content/drive/MyDrive/Rosalind/rosalind_gc (1).txt')
````

````python
data=open("/content/drive/MyDrive/Rosalind/rosalind_iprb (1).txt").read().strip()

data=data+" " #Despues de matarme la cabeza cuarenta minutos me di cuenta que
#había una solución menos elegante pero mucho más sencilla para hacer
#que funcionara el ciclo
dom=[]
cu=[]
num=[]
count=0
while count<len(data):
  for space in data:
    if space!=" ":
      cu.append(space)
      count+=1
    elif space==" ":
      dom="".join(cu)
      num.append(int(dom))
      cu=[]
      count+=1
k=int(num[0])
m=int(num[1])
n=int(num[2])

#Ahora comenzaremso con las probabilidades
poblacion= k+m+n
combinaciones= poblacion*(poblacion-1)

# 1. Ambos padres son recesivos (n * (n - 1)) -> 100% de hijos recesivos
rec_rec = (n * (n - 1)) / combinaciones * 1.0

# 2. Un padre heterocigoto y otro recesivo (m * n o n * m) -> 50% de hijos recesivos
# Multiplicamos por 2 porque el orden de elección importa (m luego n, o n luego m)
het_rec = (2 * m * n) / combinaciones * 0.5

# 3. Ambos padres son heterocigotos (m * (m - 1)) -> 25% de hijos recesivos
het_het = (m * (m - 1)) / combinaciones * 0.25

# Probabilidades de obtener un hijo recesivo
total_recesivo = rec_rec + het_rec + het_het
total_dominante = 1-total_recesivo
print(total_dominante)

#FUNCIONÓ AHHHHHH :DDDDD
````

````python
def process_fasta_and_find_consensus(fasta_string):
    # Step 1: Parse the FASTA string data line by line
    dna_strings = []
    current_dna = []

    # Split the raw text by newlines and clean whitespace
    lines = fasta_string.strip().split('\n')

    for line in lines:
        line = line.strip()
        if not line:
            continue
        if line.startswith('>'):
            # Save the previous sequence if it exists
            if current_dna:
                dna_strings.append("".join(current_dna))
                current_dna = []
        else:
            # Append sequence segments (handles multi-line DNA)
            current_dna.append(line)

    # Add the last sequence to the list
    if current_dna:
        dna_strings.append("".join(current_dna))

    if not dna_strings:
        print("No DNA sequences found.")
        return
````
