```python
#ROSALIND STRING ALGORITHMS 4

#Computing GC Content

#Instructions:
#The GC-content of a DNA string is given by the percentage of symbols in
#the string that are 'C' or 'G'. For example, the GC-content of
#"AGCTATAG" is 37.5%. Note that the reverse complement of any DNA string
#has the same GC-content.

#DNA strings must be labeled when they are consolidated into a database.
#A commonly used method of string labeling is called FASTA format.
#In this format, the string is introduced by a line that begins
#with '>', followed by some labeling information.
#Subsequent lines contain the string itself; the first line to begin
#with '>' indicates the label of the next string.

#In Rosalind's implementation, a string in FASTA format will be labeled
#by the ID "Rosalind_xxxx", where "xxxx" denotes a four-digit code
#between 0000 and 9999.

#Given: At most 10 DNA strings in FASTA format
#(of length at most 1 kbp each).

#Return: The ID of the string having the highest GC-content,
#followed by the GC-content of that string.
#Rosalind allows for a default error of 0.001 in all decimal
#answers unless otherwise stated; please see the note on absolute
#error.
```

```python
#Started creating a new space, a dictionary
fasta={}
data=open("/content/drive/MyDrive/Rosalind/rosalind_gc (1).txt").read().strip()

#To clean and separate the data I separated the sequences

#Thought process:
#1. Recognise the start of every new sequence
#2. Separate ID fasta of sequence
#3. Calculate GC content of the sequence
#4. Use dictionary: ID as key and GC as value
#5. Calculate maximum


#1. Recognise the start of every new sequence: As ">", if you
#use \n the programm ignores the possibility of multiple lines

identities=[]
sequences=[]
i=-1

for position in data:
  i=i+1

#Defined the start and end of every ID in "Rosalind_xxxx"

  if position==">":
    start=i+1
    end=i+14
    ID=data[start:end] #"Rosalind_xxxx"
    a=str(ID) #To use as key for FASTA dictionary


#2. Separate ID fasta of sequence

#Properties of phyton lists: Ordered, mutable.
# append() => Adds an item to the end of the list.
#I created and list with the IDs in the same position as the
#original string

    identities.append(a)

#Obtain the sequence, add it to the sequences list in order
#eliminate next sequence ">" to facilitate GC obtention

    adn=[]
    for position in range(end, len(data)):
        adn.append(data[position])
        if data[position]==">":
          adn.remove(">")
          break

    sec="".join(adn)
    sequences.append(sec)

#What do I have here: Separated lists in the same order for
#sequences and IDs


#3. Calculate GC content of the sequence

for cu in sequences: #cu=every single sequence for direction and GC
  a=cu.replace("\n", "") #.strip() DOES NOT WORK HERE

  #Ecuation given by Rosalind, it's a % use *100
  #count() will count the amout of times a substring appears in a string

  GC=((cu.count("C")+cu.count("G"))/len(a))*100

  #4. Use dictionary: ID as key and GC as value

  for ID in identities:
    if cu==sequences[identities.index(ID)]:
      fasta[ID]=GC


#5. Calculate maximum

for key, value in fasta.items():
  if value==max(fasta.values()):
    print(f"{key}\n{value}")
```

```python
def solve_gc_content(filename):
    with open(filename, 'r') as file:
        lines = file.read().splitlines()

    seq_dict = {}
    current_id = ""

    # 1. Parse FASTA file
    for line in lines:
        if line.startswith(">"):
            current_id = line[1:]
            seq_dict[current_id] = ""
        else:
            seq_dict[current_id] += line

    max_id = ""
    max_gc = -1.0

    # 2. Calculate GC content
    for seq_id, seq in seq_dict.items():
        if not seq:
            continue
        # Formula: (Count of G and C / Total length) * 100
        gc_count = seq.count('G') + seq.count('C')
        gc_percent = (gc_count / len(seq)) * 100

        # 3. Find the highest
        if gc_percent > max_gc:
            max_gc = gc_percent
            max_id = seq_id

    print(max_id)
    print(f"{max_gc:.6f}")

# Call the function with your input file
solve_gc_content('/content/drive/MyDrive/Rosalind/rosalind_gc (1).txt')
```
