```python
#ROSALIND STRING ALGORITHMS 2

#Transcribing DNA into RNA

#Instructions:
#An RNA string is a string formed from the alphabet containing 'A', 'C', 'G', and 'U'.

#Given a DNA string t corresponding to a coding strand, its transcribed RNA string u
#is formed by replacing all occurrences of 'T' in t with 'U' in u.

#Given: A DNA string t having length at most 1000 nt.
#Return: The transcribed RNA string of t.
```

```python
#Fisrt we read the complete content of the file with the command read()
#We clean the data with the command strip(), this command removes leading and trailing whitespace (spaces, tabs, and newlines) from a string by default.
#type(dna)=string
dna=open('/content/drive/MyDrive/Rosalind/rosalind_rna.txt').read().strip()

#For the variable rna I utilized the command replace(), this command for type string retrives a copy of the string where all occurrences of a specified substring are replaced with another substring.
#understand st.replace("a","b") as: Every "a" in the string st is replaced by a "b".
#.replace("a","b",number) will make the replacements from left to right only in the first number of times the "a" appears.

rna=dna.replace('T','U')

print (rna)
```
